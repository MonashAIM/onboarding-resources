# 2.2) X Ray <a id="xray"></a>
X rays are a medical imaging technique used to diagnose various diseases and injuries. They utilise ionising radiation, which passes beams through the human body to capture images of bones, soft tissue and underlying structures. It is an invaluable tool clinicians use frequently as it is cheap, accessible and an accurate diagnostic tool. 

### Types 
There are various types of X rays, all of which work slightly differently to capture different images. Some of these include 
Radiography: This imaging technique is used to identify broken bones, dental issues, and chest conditions. It uses the least amount of radiation and is the most simple form of X-ray imaging. 
Fluoroscopy: This imaging method observes a patient's real-time movements and captures snapshots. For example, it is used to monitor gut activity after a barium meal. Barium is a contrast dye which is radio opaque. When the patient ingests the die, clinicians can capture how the dye moves through the Gastrointestinal system. 
Computed Tomography (CT): A CT machine constructs a 3D image of a person through various "slices." This technology is highly accurate in identifying tumours and changes in muscle and bone structure. In order to create such an accurate image, it uses the highest amount of radiation. 
Mammography: This is a specialised X ray that is used for screening for breast cancer. 

### Physics 
Essentially, in X ray imaging, the X ray beams travel through the body and are absorbed by the various structures within the body such as bone and soft tissue. All the structures within the body vary in density and as such, the amount of radiation absorbed by each structure will differ, thus creating an image. For instance, bones are very radiologically dense in comparison to soft tissue structures. As such, bones appear a very solid white on the X ray, whereas soft tissue like muscles and air cavities appear in various shades of grey. 




For basic radiography, the patient is positioned in between an X ray source and an x ray detector. As the machine is turned on, the beams will pass through the patient. 



### Common Uses and Findings of Radiography 
__Bones and teeth:__ It is used in cases of trauma as it can identify fractures all over the body. It can also be used to detect arthritis of bones, bone cancers and dental decay. 

__Chest X ray:__ It can detect lung infections, a collapsed lung, fluid in the lung, a mass and other lung pathologies. It can also detect an enlarged heart and any rib fractures. 

__Abdominal X ray:__ It can detect digestive tract issues such as a blocked bowel, perforation of bowel and swallowed items. It can also visualise masses and certain types of stones. 

### Advantages / Disadvantages 
__Advantages:__ 

Low risk, non invasive, fast and accurate. They can also assist medical professionals in performing procedures on patients with greater accuracy. 

__Disadvantages:__  

However, to function, they emit very low levels of ionising radiation, which according to the WHO, is a carcinogen. Although the risk is minimal, the long-term effects of ionising radiation from X-rays can include causing cancer many decades or years later. For instance, the radiation emitted from one chest X-ray is equivalent to 2.4 days of natural background radiation, while a CT scan of the abdomen is equivalent to 2.7 years of natural background radiation.

### Current Progress on AI in Radiography 
The utilisation of AI in X-ray imaging represents a heavily researched domain. Numerous studies have been conducted to assess its efficacy in detecting abnormalities in imaging, with the primary objective being a slight increase in automation to save radiologists time, thereby enabling them to focus on other aspects of patient care.
One notably recent study utilised an AI tool to identify abnormal chest X-rays, classifying the images as either "high-confidence normal" or "not high-confidence normal." Achieving a sensitivity of 99.1%, when compared against three radiologists used as a reference standard, this marks an incredible leap forward. 

In another study involving lung X-rays and radiologists, it was found that the radiologists outperformed the AI technology. However, this discrepancy is thought to stem from the radiologists' ability to synthesise information from patients' clinical history and previous imaging studies. If AI systems could mimic this ability, they would significantly enhance their capabilities.

Thus far, AI in medical imaging continues to be under scrutiny, with clinicians consistently outperforming these technologies. Therefore, they should not be solely relied upon for making diagnoses autonomously. However, the next generation of AI systems holds promise in significantly reducing the workload for radiologists.





# 2.3) Computed Tomography (CT) <a id="ct"></a>
Computed Tomography (CT) is a medical imaging technique that uses a
combination of X-rays and computer processing to create cross-sectional
images of the body, providing detailed information on the bones, blood vessels,
and soft tissues.
 
### Physics Principles
X-rays are a form of light with the unique property that allow it to pass through
through soft tissue but not denser materials like bone. For the first time, bones
and foreign bodies can be visualised without the need for invasive surgeries.
However, soft tissues, such as brains, are still an amorphous blob



Hounsfield, the electrical engineer who invented CT, attempted to solve this
through utilising multiple X-rays beams that is later compiled by a computer
algorithm that collates the information.
During a CT scan, the X-ray tube and detectors rotate around the patient, who
is positioned at the centre of the rotation. 



Each rotation produces a single "slice" or cross-sectional image of the body,
and the process can be repeated to produce multiple slices at different levels.



The strength of each beam is detected on the opposite side. The denser the
material, the lower proportion of the original ray will be detected. Therefore, a
stronger signal indicates theyd traveled through less dense material.



Finally, computer algorithms are used to reconstruct the 2D cross-sectional
images.



The first CT, despite the low resolution, was able to identify a
cystic mass in the frontal lobe. It would not have been visible on the Xray, and certainly cannot be palpated on examination.



### Advantages & Disadvantages
__Advantages:__

1. High Resolution: CT scans offer high-resolution images, enabling detailed
visualization of the internal structures of the body. This clarity is crucial for
diagnosing diseases and injuries.
2. Speed: A CT scan can be completed within a few minutes, providing rapid
results that are essential in emergency situations.
3. Versatility: Useful for examining a wide range of body parts, including
bones, muscles, fat, organs, and blood vessels.
4. Guided Procedures: CT imaging can guide certain diagnostic procedures,
such as biopsies, reducing the risk of injury to surrounding structures.

__Disadvantages:__

1. __Radiation Exposure:__ Patients are exposed to ionizing radiation, which,
although generally low, can contribute to a small increased risk of cancer.
2. __Contrast-Related Risks:__ The use of contrast agents (dye) can cause
allergic reactions or affect kidney function, especially in patients with preexisting kidney issues.

3. __Cost:__ CT scans can be expensive, depending on the healthcare system and
whether insurance covers the procedure.
4. __Limited Soft Tissue Contrast:__ While advanced, CT scans may not always
provide as clear of a distinction in soft tissues compared to modalities like
MRI.

### Most Relevant Clinical Presentations That Require the Scan
CTs are not only known for their diagnostic value, but are also important in
ruling out other potential diagnoses. 

__Some common examples include:__
1. __Head Injuries:__ To detect fractures, bleeding (haemorrhage), brain injury
and stroke.
2. __Chest Pain:__ To examine the lungs for conditions such as pneumonia,
pulmonary embolism, or tumours, and the heart for assessing coronary
artery disease.
3. __Abdominal Pain:__ To identify appendicitis, diverticulitis, kidney stones, or
liver diseases.
4. __Cancer Diagnosis and Monitoring:__ To detect various cancers, monitor
treatment progress, and check for metastasis.
5. __Trauma:__ To rapidly assess for internal organ injuries and locate sites of
haemorrhage

### Common Findings
1. __Tumors:__ Both benign and malignant tumors can be identified, with details
regarding their size, location, and effect on surrounding tissues.
2. __Infections:__ Signs of infections, such as abscesses and swelling, can be
spotted in various organs.
3. __Vascular Diseases:__ CT angiography, where contrast is injected to highlight
blood vessels, can reveal blockages, aneurysms, and dissections.
4. __Fractures and Bone Diseases:__ Detailed images of bone injuries and
pathologies like osteoporosis.

### Current Progress of AI Integration
Perhaps the most impressive AI integration to CT scans are that of CT brains
(CTB). [annalise.ai](https://annalise.ai/annalise-reporting/), an Australia based medtech company, is capable of generating complete reports for radiologists to sign off. According to their
website, it is already Generating complete draft reports for __35% of CXR and CTB cases__ already today.

Other major advancements that are mostly in the research stage currently include:

1. __Image Reconstruction:__ AI algorithms are used to reconstruct high-quality
images from less data, reducing scan time and patient exposure to
radiation.
2. __Automated Detection:__ AI can assist in identifying and quantifying
abnormalities such as tumors, fractures, and vascular diseases, speeding
up diagnosis and reducing the potential for human error.
3. __Predictive Analysis:__ AI models can predict the progression of diseases by
analyzing CT images over time, aiding in treatment planning and
prognostication.



# 2.4) Magnetic Resonance Imaging (MRI) <a id="mri"></a>
Magnetic Resonance Imaging (MRI) is an imaging technique that creates high-definition two-dimensional and three-dimensional images of a patients body using magnetic fields and radio waves. MRI has a distinct advantage compared to X-rays, CTs and PET scans as it uses non-ionising radiation. It also provides superior contrast for soft tissues in comparison to these other imaging modalities. 

### Part One: The Physics

Hydrogen nuclei are abundant in our body in water, fat and carbohydrates. MRI positions the density and environment of these hydrogen nuclei to create images, leveraging fundamental physics and properties of protons. 

All subatomic particles possess a property called spin, which is a form of angular momentum - much like the angular momentum of a spinning basketball on someones finger. Spin was discovered after we realised that subatomic particles exhibit magnetic properties, consistent with a spinning electric charge. However, the speeds that subatomic particles would have to spin at in order to be consistent with measurements would exceed the speed of light if this was physically the case. Instead, spin is a quantum property which follows probabilistic laws and exists in quantized (specific) states. It is described as intrinsic angular momentum, you cant make it stop, slow it down or speed it up like you could with a spinning basketball.

The spin of hydrogen makes it magnetic along its rotational axis. The magnetic moment is normally random but under a more powerful external magnetic field, the magnetic moment and spin will align in one of two ways - parallel or antiparallel. Parallel refers to alignment in the direction of the external magnetic field while antiparallel refers to alignment in the opposite direction. Low energy hydrogen nuclei (protons) will align parallel and higher energy protons will align anti-parallel. We expect the majority of protons to align parallel in a human body. This produces a net magnetization vector in the same direction of the external magnetic field. We cant measure this directly, so we need to influence its behaviour. 

The magnetic moment will precess around the direction of the external magnetic field at the Larmor frequency. If we emit a radio wave at the same frequency as the Larmor frequency, were able to flip parallel protons antiparallel to cancel the net magnetization vector in the direction of the magnetic field and induce a net transverse magnetization at a right angle to the direction of the magnetic field. Note these are net statistical changes, not every proton is in exact alignment with the transverse plane, instead cancelling each other out to produce this net effect. This transverse state is temporary, and the protons are inclined to return to their original state of magnetic alignment and spin. These protons precess along the new temporary transverse axis of magnetization at the Larmor frequency and induce currents in the receiving coils of the MRI machine. The time it takes for the net magnetization vector to return to its original alignment (from right angled to parallel to the external magnetic field) can be quantified by the T1 (longitudinal relaxation) time. It is measured after the last radiofrequency pulse has been applied, one T1 is approximately when 63% of the original net magnetization vector is recovered, two T2 is 86% of the original and full recovery occurs normally in three to five T1 times. The T2 time (transverse relaxation time) is the time it takes for the transverse magnetization signal to decay to approximately 37% of its initial magnitude. This is dependent on the chemical properties of a specific tissue, T2 is long in water as its hydrogens are highly energetic but much slower in other tissues like cerebrospinal fluid and fatty tissues. T1 relaxation times are always longer or equal to T2 relaxation times, and represents a transfer of energy (which is why its also called spin-lattice relaxation, with lattice referring to the external structures it gives energy to).

By using T1 and T2 relaxation times, repetition time (TR) and echo time (TE), gradient electric coils to localise tissues (to map spatial changes, as the local magnetic field strength is directly proportional to the Larmor frequency) and pulse sequences (that preference certain T1, T2 relaxation times to attenuate different tissue properties and thus types) were able to get an output MRI signal with spatial frequencies called the K-space. The K-space is transformed via Fourier transformation into the final image space. Final contrast is determined by proton density, T1 and T2 relaxation times and can be changed based on  T1, T2 weighted techniques such as introduction of contrast agents (e.g. Gadolinium) or using different pulse sequences. The signal-to-noise ratio is also a factor, and technicians change TR and TE values, voxel volume and receiver bandwidth to minimise this. 

### Part Two: Use Cases, Advantages and Disadvantages

Disease aetiology is normally reflected in changes of tissue position, density or size. MRI can help image these changes for both diagnosis and tracking of disease, with particular usefulness for changes in soft tissue as opposed to bone. Cancer, soft tissue injuries, joint, spinal and organ injuries and diseases can all be diagnosed and tracked through MRI imaging. [1] The usage of contrast agents can be used to attenuate different tissue types, most commonly gadolinium administered intravenously for T1 weighted images. This allows greater clarity in MRI for angiography, venography and detection of focal lesions. [2] Functional MRI is another specific use case of MRI in which patients perform cognitive tasks while in the MRI machine, so that changes in brain activity (perfusion) can be measured. This maps brain activity and function and is particularly useful in planning brain surgeries and assessing brain damage. [3]

__Advantages:__ 
1. it uses no ionising radiation.
2. is non-invasive .
3. offers superior soft tissue contrast to other imaging modalities.
4. is able to create multiple planes as well as 3D construction of images.
5. is able to perform tissue characterisation.

__Disadvantages:__ 
1. It's price for purchase and operation.
2. Need for trained technical team for operation.
3. Possibility of patient discomfort due to the small space with loud noises.
4. The inability of MRI to be used on patients with embedded or internalised ferromagnetic materials. [4] 


### Part Three: Current state with regards to Artificial Intelligence (AI)

MRI science is a highly technical field and has offered a significant potential space for the integration of AI, including deep learning, technology. Current use cases and frontiers for development are: 1) for programming and planning MRI scans without a specialist - e.g. cardiac MRI 2) automatic and early detection of lesions or abnormal changes in physiology - e.g prodromal Alzeheimers [5] 3) ability to stage diseases through tissue characterisation 4) upscaling images, improving clarity, reducing scan times and improving overall performance. # Positron Emission Tomography (PET) <a id="pet"></a>
_The use of a radioactive compound to make three-dimensional images of organs, tissues and blood flow in the body. This is a form of nuclear medicine imaging._ <br><br> 
Authored by Thomas Axnick for Monash AIM. 21/02/2024.

### Basic physics principles 
A small amount of a radioactive tracer is inserted into a vein, which consists of a positron-emitting isotope (radioactive tag) bound to an organic ligand (targeting agent). This allows the tracer to bind specifically to the tissue being targeted.
After the tracer is bound to the target, the positron collides with an electron in the tissue, creating two photons in the process known as annihilation. As these photons are emitted, they are detected by a special camera, which can then represent the data as a 3D image. 


[1] An image describing how PET scans work, using Fluorine-18 as an example. <br>
Some of the common isotopes and their uses are below (examples present in Common Findings):
__Fluorodeoxyglucose (FDG):__ <br>A glucose analogue. As tissue in the body use glucose as energy to function, areas with high glucose turnover (such as the brain and liver) will accumulate the radioactive iodine. As cancers are cells with abnormally high metabolism, their glucose use will be increased, and can be picked up on PET scans.

__Prostate-Specific Membrane Antigen (PSMA):__ <br>An antigen for a receptor that has high expression on prostate cancer cells compared to normal prostate tissue. As the cancer cells have a higher number of receptors, they will attach more PSMA compared to healthy tissue, and will thus be more radioactive. 

__Fluorodopa (FDOPA):__ <br>A radioactive L-DOPA analogue, which is a precursor for dopamine. In a disease such as Parkinsons Disease, there is a loss of nerves that use dopamine (dopaminergic neurons) and so using a PET scan will show less activity/radioactivity as compared to someone without Parkinsons. 


##### __Advantages:__
The greatest of PET scans is the ability to detect changes at the cellular level. Many other scan mediums cannot demonstrate how a cell interacts with molecules, only shape/size/location of tissue, which can be done with PET scans anyways. Due to these, PET scans are a great modality for follow-up appointments to track progress of cancer growth or spread. <br>

PET scans can be used for many, many conditions due to the customisability of the tracer. Stated below, it can be used from anything from cancer care to heart check-ups. <br>

PET scans are also great at detecting distant metastases  other scans are for more specific regions, and are usually not whole-body due to time and cost. It is also on the faster end compared to CT and MRI, but slower than X-ray and ultrasound. 

##### __Disadvantages:__
There are __disadvantages__ of PET scans though. The radiation exposure can increase the risk of cancer in children, or when doing multiple radioactive imaging modalities. <br>

PET scans also have low resolution compared to CT and MRI which can increase difficulties localising masses when near other anatomical structures. To combat this, PET can be combined with MRI to get the best of both. 

Also, for those with claustrophobia, PET scans can pose a challenge. As patients need to stay still throughout the procedure, and the machine is essentially a tube, radiographers need to take special care in keeping the patient still as movement results in a lower resolution image. 

### Most relevant clinical presentations that require the scan:
Oncology (cancer medicine) is a major part of PET scan use; it is both used to detect a cancer but also for follow-up evaluations to see how it is progressing, to see how the cancer is reacting to treatment, and to see if there is any spread to other tissues or organs. 
It can also be used for neuroscience, using many types of radiotracer; analogues of glucose, oxygen, and/or L-DOPA (the precursor for dopamine) can be used to see neurological damage to the brain. Conditions such as Alzheimers disease and Parkinsons disease can be seen using PET scans. 
Other uses include in cardiology (how blood flows through the heart), infectious disease (localisation of infective source  for example after joint replacements or bladder infections), and autoimmune conditions (where in the body is affected and to what extent). 

### Common findings: 


A focal uptake of PSMA, or prostate-specific membrane antigen, in the prostate. There are other suspicious lesions present within the area but much harder to see. Note how the salivary glands in the face are lit up; the PSMA tracer is not quite 100% specific for the prostate tissue.

A normal vs abnormal PET scan using FDOPA. The brain scan on the right shows reduced uptake globally, suggesting Parkinsons Disease. There is a large concentration of neurons in the caudate and putamen (collectively called the striatum) which shows a larger radioactivity compared to the rest of the brain, but again the abnormal scan shows markedly lower activity.

### Current progress of AI integration:
AI is getting used more and more in the medical field. I am unsure if any medical companies use AI for radiology for diagnosis as of early 2024, but there are increasing numbers of studies showing how useful it could be. <br>
A study showed 77% detection of lung cancer over 3,000 cases. <br>
Another study showed a 90% accuracy in detecting a lung cancer.<br>
One study showed 95.9% sensitivity and 98.1% specificity in detecting lung cancer.<br>
Additional studies showed an AUC (accuracy) of 0.85, compared to 0.61 and 0.60 of a radiologist and pulmonologist, respectively. <br>
A 2015 study had a false positive rate of 72.8%, which with recent advancements has decreased to 4.9%. 
AI in the medical field has issues in terms of liability; if a patient has a misdiagnosis from the AI, who is to blame? To combat this, AI can be used in conjunction with a clinical decision as more of a fail-safe, but then radiologists may still use less caution in making a diagnosis as if they miss something, it will get picked up anyways. Ethics and legality need to be advanced fast to catch up with this evolving field of medicine. 




# Echocardiogram <a id="echo"></a>
An Echocardiogram (echo) is a medical imaging technique used mostly to look at
structures within and surrounding the heart (i.e. atria and ventricular size, cardiac
valves, adjacent vessels, etc). An echocardiogram is a subset of an ultrasound,
using sound waves to produce a live image.

### Principles:
Ultrasound simply refers to sound waves with a frequency outside of human
hearing (i.e. >20kHz), typically around 4-7MHz for an echocardiogram. These
sound waves have the ability to pass through and interact with human tissues,
reflecting and scattering. The interaction (and therefore scattering or reflection) of
the sound waves is dictated by the composition and density of the material,
associated with a property called acoustic impedance which is unique to each
tissue type. We can measure these outputs and therefore construct an image in
real time based on the readings.



### Echocardiogram Types and Setup:
There are 2 main modalities for an echocardiogram used on adults; the
Transthoracic Echocardiogram (TTE) or Transoesophageal Echocardiogram (TOE).
A TTE is performed by pressing the transducer (which produces and receives the
ultrasound waves) onto a bare chest, a non-invasive method. A TOE requires
passing the transducer down the oesophagus, an invasive procedure, but yields
more accurate and detailed images.



The machine used for echocardiography is very similar to the machine used for
other types of ultrasound, often composed of a transducer (which produces and
receives ultrasound waves), an operating monitor and panel unit.



Whilst there are different types of echocardiograms, there are also different
methods of using an echocardiogram. Echos can be constructed to provide a twodimensional or three-dimensional image. Doppler echocardiograms are also
available, which are utilised to measure the direction of blood flow within the
heart.



### __Advantages:__
1. __Non-Invasive:__ <br>Transthoracic Echocardiography is a non-invasive procedure,
therefore not requiring insertion of instruments into the body, decreasing the
risks associated with its use.
2. __Real-Time Imaging:__ <br>Echocardiography provides real-time images of the heart,
allowing observation of the hearts structure and function, particularly with
blood flow and valvular abnormalities.
3. __Versatility:__ <br>Echocardiography has a wide variety of modalities and methods,
allowing for a comprehensive evaluation of the hearts anatomy.
4. __No Radiation Exposure:__ <br>Echocardiograms use ultrasound waves to produce
their images, therefore not requiring exposure to potentially ionising forms of
radiation such as X-rays.

### __Disadvantages:__
1. __Operator-Dependent:__ <br>The quality of echocardiographic images is highly
dependent on the skill and experience of the operator, and therefore crucial
details can be missed.
2. __Limited Suitability:__ <br>Some patients may have various factors (such as obesity
or chest wall abnormalities) that prevent a clear picture of the heart from being
obtained.
3. __Limited Penetration:__ <br>Whilst a transthoracic echocardiogram is often suitable
for superficial structures, when observing deeper anatomy (i.e. the aortic
valve or descending aorta) a transoesophageal echocardiogram is often
required.
4. __Visualisation:__ <br>An echocardiogram may not provide adequate visualisation of
certain cardiac structures or abnormalities, particularly those located behind
bone or air-filled structures, requiring CTs or MRIs for further evaluation.

### Conditions Requiring Echocardiography
__Structural Abnormalities:__ <br>Echocardiography is useful for identifying
structures within the heart, such as congenital heart defects, valvular
abnormalities and heart wall abnormalities. This correlates with conditions
such as cardiomyopathies, valvular defects, and septal defects.



__Evaluation of Cardiac Function:__<br>Cardiac function is often related to ventricular
size and contractility. Parameters such as ejection fraction and global
longitudinal strain can assess the hearts pumping ability. This is useful for
detecting conditions such as heart failure, myocardial infarction and
cardiomyopathies.



__Assessment of Valve Structure:__ <br>Echocardiography can also view the valves
within the heart. This is particularly useful for valve stenosis, regurgitation or
prolapse (such as aortic stenosis or mitral prolapse).



__Pericardial Abnormalities:__ <br>Echocardiography also provides a useful view of
the hearts sac, or pericardium. This is particularly useful to identify a
pericardial effusion, which may be linked with pericarditis or cardiac
tamponade.



__Evaluation of Cardiac Chambers and Blood Flow:__ <br>The size, shape and
movement of the hearts chambers (the atria and ventricles) can also be
monitored with echocardiography. Doppler echocardiography is particularly
useful for evaluating blood flow patterns and velocities.



### Current Progress with Artificial Intelligence:
Currently, artificial intelligence is already used with echocardiography, particularly
for automated structural and functional measurements.
Future development is focussed on its ability to identify specific disease states
within the heart, particularly defects which may be too nuanced or small for a
cardiologist or other interpreter to identify. This has profound implications, as
often the interpretation of an echocardiogram is imperfect, leaving significant
room for AI optimisation.
